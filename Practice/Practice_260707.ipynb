```python
#23.7
row, col = map(int, input().split())

matrix = []
for i in range(row):
    matrix.append(list(input()))

for r in range(row):
    for c in range(col):
        if matrix[r][c] == '*':
            continue
        count = 0
        for nr in range(r-1, r+2):
            for nc in range(c-1, c+2):
                if 0 <= nr < row and 0 <= nc < col:
                    if matrix[nr][nc] == '*':
                        count += 1
        matrix[r][c] = str(count)
for r in range(row):
    print(''.join(matrix[r]))
```

```python
#31.4
def is_palindrome(word):
    if len(word) < 2:
        return True
    if word[0] != word[-1]:
        return False
    return is_palindrome(word[1:-1])
print(is_palindrome('level'))
```

```python
#31.5
def fibonacci(n):
    a = 0
    b = 1
    for i in range(n):
        a, b = b, a + b
    return a
n = int(input())
print(fibonacci(n))
```

```python
# 위치 인수의 리스트 언패킹
#3 위치 인수를 사용하는 함수

def print_numbers(a, b, c):
    print(a)
    print(b)
    print(c)
```

```python
print_numbers(10, 20, 30)
```

```python
## 언패킹 사용하기

x = [10, 20, 30]
print(x)
print(*x)   # 언패킹돼서 인수로 나옴
```

```python
## 가변 인수 함수 만들기
def print_numbers(*args):    # 매개변수 수를 정하지 않고 입력되는 대로 받는다는 뜻
    print(args)     
    for arg in args:
        print(arg)

print_numbers(10)
print_numbers(10, 20, 30, 40)

print()

x = [10]
y = [10, 20, 30, 40]

print_numbers(*x)
print_numbers(*y)
```

```python
# 키워드 인수 사용하기
def personal_info(name, age, address):
    print('이름: ', name)
    print('나이: ', age)
    print('주소: ', address)

personal_info('홍길동', 30, '서울시 용산구 이촌동')

print()

personal_info(name='홍길동', address='서울시 용산구 이촌동', age=30)
# personal_info(age=30, name='홍길동', '서울시 용산구 이촌동') # 요거는 안됨. 앞에 키워드 인수를 썼으면 뒤에도 다 키워드 인수를 써야됨
personal_info('홍길동', address='서울', age = 30)    # 요거는 됨
```

```python
# 키워드 인수와 딕셔너리 언패킹 사용하기
def personal_info(name, age, address):
    print('이름: ', name)
    print('나이: ', age)
    print('주소: ', address)

x = {
    'name': '홍길동',
    'age': 30,
    'address': '서울시 용산구'
}

personal_info(**x)
# personal_info(name='홍길동', age=30, address='서울시 용산구 이촌동')   # 요거와 같음
```

```python
# 키워드 인수를 사용하는 가변 인수 함수 만들기

def personal_info(**kwargs): 
#     print(kwargs)
#     print(type(kwargs))
    for kw, arg in kwargs.items():
        print(kw, ': ', arg, sep='')

personal_info(name='홍길동', age=30)
```

```python
x = {
    'name': '홍길동'
}

y = {
    'name': '홍길동',
    'age': 30,
    'address': '서울시 용산구'
}

personal_info(**x)
personal_info(**y)
```

```python
# 매개변수에 초깃값 지정하기
def personal_info(name, age, address='비공개'):
    print('이름: ', name)
    print('나이: ', age)
    print('주소: ', address)

print(personal_info('홍길동', 30))    # address가 비공개로 디폴트값이 설정되어 있어서 인수를 두 개만 줘도 실행이 됨
print(personal_info('홍길동', 30, '서울시 용산구'))    # 기본적으로 디폴트값이 있지만 지정해주면 해당 데이터로 덮어씌워짐
```

```python
#30.7
korean, english, mathematics, science = map(int, input().split())

def get_max_score(*agrs):
    return max(agrs)

def get_min_score(*agrs):
    return min(agrs)

def get_avg_score(*agrs):
    return (sum(agrs) / len(agrs))

max_score = get_max_score(korean, english, mathematics, science)
min_score = get_min_score(korean, english, mathematics, science)
avg_score = get_avg_score(korean, english, mathematics, science)

print(f"낮은 점수: {min_score}, 높은 점수: {max_score}, 평균 점수: {avg_score}")


max_score = get_max_score(english, mathematics)
min_score = get_min_score(english, mathematics)
avg_score = get_avg_score(english, mathematics)

print(f"영어 수학 중 낮은 점수: {min_score}, 높은 점수: {max_score}, 평균 점수: {avg_score}")
```

```python
# 세트만들기
fruits = {'strawbeery', 'grape', 'orange', 'pineapple', 'cheery'}
print(fruits)
print(type(fruits))
```

```python
# 세트에 특정 값이 있는 지 확인하기
print('orange' in fruits)
print('peach' in fruits)
```

```python
# set을 사용하여 세트 만들기
a = set('apple')    # 중복된 문자는 포함하지 않음
print(a)

b = set(range(5))
print(b)

c = set()
print(c)

d = {}   # set() 를 안하고 비우면 딕셔너리가 됨
type(d)
```

```python
# 집합 연산 사용하기
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print(a | b)      # union , 합집합
print(a.union(b))
print(set.union(a, b))
print()

print(a & b)     # intersection, 교집합
print(a.intersection(b))
print(set.intersection(a, b))
print()

print(a - b)     # difference, 차집합
print(a.difference(b))
print(set.difference(a, b))
print()

## 대칭 차집합 = 합집합에서 교집합을 뺀 나머지 부분
print(a ^ b)
print(set.symmetric_difference(a, b))
```

```python
a = {1, 2, 3, 4}
b = a
print(id(a))
print(id(b))

b.add(5)    # a와 b는 같으므로 b가 바뀌면 a도 바뀜
print(a)
print(b)
```

```python
a = {1, 2, 3, 4}
b = a.copy()
print(id(a))
print(id(b))

b.add(5)        # b는 a를 복사한 것이므로 값은 같지만 다른 자료기에 b가 바뀌더라도 a는 바뀌지 않음
print(a)
print(b)
```

```python
# 반복문으로 세트의 요소를 모두 출력하기
a = {1, 2, 3, 4}

for i in a:
    print(i)
```

```python
# 세트 컴프리헨션
# {식 for 변수 in 반복가능한 객체}
# set(식 for 변수 in 반복가능한 객체)

a = {i for i in "apple"}
print(a)
```

```python
# 세트 표현식에 if 조건문 사용하기
# {식 for in 세트 if 조건식}
# set(식 for in 세트 if 조건식)

a = {i for i in "pineapple" if i not in "apl"}    # a, p, l 빼고 나머지를 집합 자료형으로 출력
print(a) 
```

```python
#23.9.Q1
num1, num2 = map(int, input().split())

div1 = { i for i in range(1, num1 +1) if num1 % i == 0}
div2 = { i for i in range(1, num2 +1) if num2 % i == 0}

divisor = div1 & div2
print(sum(divisor))
```

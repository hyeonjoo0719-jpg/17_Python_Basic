```python
# 변수만들기
##1. 영문문자, 숫자를 사용
##2. 대소문자를 구분
##3. 문자부터 시작을 해야하고, 숫자부터 시작은 안됨 a1(o) 1a(x)
##4. 언더바로는 시작할 수 있음
##5. 특수문자(+, -, *, /, ...)는 사용할 수 없음
##6. 파이썬 키워드(if, for, while, print 등)는 사용할 수 없음
```

```python
# 변수 여러 개를 한번에 만들기

x, y, z = 10, 20, 30

print(x)
print(y)
print(z)
```

```python
x = y = z = 10
print(x)
print(y)
print(z)
```

```python
x, y = 10, 20
x, y = y, x
print(x)
print(y)
```

```python
# 변수로 계산하기
a = 10
b = 20
print(a + b)
```

```python
# 산술 연산 후 할당 연산자 사용하기
a = 10
a + 20
print(a)
```

```python
b = 10
b += 20
print(b)
```

```python
a, b = map(int, input("숫자 두 개를 입력하세요: ").split())
print(a)
print(b)
print(a + b)
```

```python
# 입력받은 값을 콤마를 기분으로 분리하기
a, b = map(int, input("숫자 두 개를 입력하세요: ").split(','))
print(a)
print(b)
print(a + b)
```

```python
print("10,20".split())
print("10,20".split(','))
```

```python
a, b = map(int, input("숫자 두 개를 입력하세요: ").split())   # list(map(int, input("숫자 두 개를 입력하세요: ").split())) 와 같음
print(a + b)
```

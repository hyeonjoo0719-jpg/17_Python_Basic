```python
# 딕셔너리 만들기
# 딕셔너리 = {}
#     키1: 값1,
#     키2: 값2,
#     키3: 값3,
#     ...
# }
lux = {
    "health": 490,
    "mana": 334,
    "melle":550,
    "armor": 18.72
}
```

```python
# 키 이름이 중복되면?
lux = {
    "health": 490,
    "health": 800,   # 나중에 입력된 값이 저장됨
    "mana": 334,
    "melle":550,
    "armor": 18.72
}
print(lux)
```

```python
# 딕셔너리 키의 자료형
x = {
    100: "hundred",
    False: 0,
    3.5: [3.5, 3.5]
}
print(x)
```

```python
x = {
    [10, 20] : 100
}
print(x)
```

```python
# 빈 딕셔너리 만들기
x = {}
y = dict()

print(x)
print(y)
print(type(x))
print(type(y))
```

```python
# dict로 딕셔너리 만들기
lux1 = dict(
    health=490,
    mana=334,
    melee=558,
    armor=18.72
)
print(lux1)
```

```python
lux2 = dict(zip(
    ['health', 'mana', 'melee', 'armor'],
    [490, 334, 550, 18.72]
))
print(lux2)
```

```python
lux3 = dict([
    ('health', 490),
    ('mana', 334),
    ('melee', 550),
    ('armor', 18.72)
])
print(lux3)
```

```python
lux4 = dict([{}
    'health', 490,
    'mana', 334,
    'melee', 550,
    'armor', 18.72
}])
print(lux4)
```

```python
# 딕셔너리의 키에 접근하고 값 할당하기
lux = {"health": 490,
    "mana": 334,
    "melle":550,
    "armor": 18.72
}

print(lux["health"])
print(lux["armor"])
```

```python
lux["health"] = 2037
lux["mana"] = 1104
print(lux)
```

```python
lux["mana_regen"] = 3.28
print(lux)
```

```python
lux["attack_speed"]  # 없는 걸 인덱싱하면 에러남
```

```python
lux = {"health": 490,
    "mana": 334,
    "melle":550,
    "armor": 18.72
}

print(lux.get("attack_speed", None))   # 없는 값을 호출하면 None 값 출력
```

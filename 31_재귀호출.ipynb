```python
# 재귀 호출 사용하기

def hello():
    print('Hello world')
    hello()

hello()
```

```python
# 재귀호출에 종료 조건 만들기

def hello(count):
    if count == 0:
        return     # 함수 종료
    print("Hello World", count)

    count -= 1
    hello(count)

hello(5)
```

```python
# 재귀호출로 팩토리얼 구하기

def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

```python
#31.4
def is_palindrome(word):
    if len(word) < 2:
        return True

    if word[0] != word[-1]:
        return False

    return is_palindrome(word[1:-1])

print(is_palindrome('hello'))
print(is_palindrome('level'))
```

```python
#31.5

def pinabochi(n):
    for i in range(n+1):
        pin = i + (i +1)
    return pin


n = int(input())
print(pinabochi(n))
```

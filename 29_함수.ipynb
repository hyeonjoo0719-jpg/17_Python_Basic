```python
# 덧셈 함수 만들기
def add(a, b):
    """
    매개변수
    a : int
    b : int

    출력: int
    """
    print(a + b)

add(3, 5)

add.__doc__   # 독스트링을 뭘 썼는 지가 나옴
```

```python
def add2(a, b):
    print(a + b)
    return

y = add2(10, 20)   # print(a + b)에 대한 출력
print(y)           # return에 대한 출력
```

```python
# 함수에서 값을 여러 개 반환하기
def add_sub(a, b):
    return a + b, a - b    # 함수는 return을 만나면 종료되기 때문에 그 다음 return은 실행 안됨. 한 로직 상에 return은 한개만 있어야 함

c = add_sub(10, 20)
print(c)      # 튜플 형태로 값을 출력

x, y = add_sub(10, 20)
print(x)     # 언패킹
print(y)     # 언패킹
```

```python
# 함수 호출 과정 보기
def mul(a, b):
    c = a * b
    return c

def add(a, b):
    c = a + b       # a와 b, c는 지역(?) 변수라서 위에 것들과 별개임
    print(c)
    d = mul(a, b)
    print(d)

x = 10
y = 20
add(x, y)
```

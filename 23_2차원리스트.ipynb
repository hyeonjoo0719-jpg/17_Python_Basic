```python
# 반복문으로 2차원 리스트의 요소 모두 출력하기
a = [
    [10, 20],
    [30, 40],
    [50, 60]
]

for x, y in a:
    print(x, y)
```

```python
# for 반복문을 두 번 사용하기
a = [
    [10, 20],
    [30, 40],
    [50, 60]
]

for i in a:      #3번 돌아감
    for j in i:         #요소가 두개라서 두번돌아감
        print(j, end=' ')
    print()
```

```python
a = [
    [10, 20],
    [30, 40],
    [50, 60]
]
for i in range(len(a)):
    for j in range(len(a[i])):
        print(a[i][j], end=' ')
    print()
```

```python
# while 반복문을 한 번 사용하기
a = [
    [10, 20],
    [30, 40],
    [50, 60]
]

i = 0

while i < len(a):
    x, y = a[i]
    print(x, y)
    i += 1
```

```python
# while 반복문을 두 번 사용하기
a = [
    [10, 20],
    [30, 40],
    [50, 60]
]

i = 0
while i < len(a):
    j = 0
    while j < len(a[i]):
        print(a[i][j], end=' ')
        j += 1
    print()
    i += 1
```

```python
#23.7.Q1
matrix = []
row, col = map(int, input().split())

for i in range(row):
    matrix.append(list(input()))


for r in range(row):
    for c in range(col):
        if matrix[r][c] == '*':
            continue

        count = 0
        for nr in range(r-1, r+2):
            for nc in range(c-1, c+2):

                if 0 <= nr < row and 0 <= nc < col:
                    if matrix[nr][nc] == '*':
                        count += 1
        matrix[r][c] = str(count)
    
for r in range(row):
    print(''.join(matrix[r]))
```

```python
row, col = map(int, input().split())
matrix = []
for i in range(row):
    matrix.append(list(input()))

for r in range(row):
    for c in range(col):
        if matrix[r][c] == '*':
            continue              # 지뢰가 있는 부분인 *은 그냥 넘어감
     
        count = 0    # 지뢰 수를 카운트 하기 위해 0으로 시작
        for nr in range(r-1, r+2):     # 0행부터 2행까지
            for nc in range(c-1, c+2):  # 0열부터 2열까지
                
                #(범위 밖은 카운트 할 필요x)  #(row를 벗어나는건 count할 필요 x) 
                if 0 <= nr < row and 0 <= nc < col:
                    if matrix[nr][nc] == '*':
                        count += 1
        matrix[r][c] = str(count)

print(matrix)
for r in range(row):
    print(''.join(matrix[r]))
```

```python
# else와 들여쓰기
x = 5

if x == 10:
    print('10입니다.')
else:
    print('x에 들어있는 숫자는')
    print('10이 아닙니다.')
```

```python
# if 조건문에 숫자 지정하기 (숫자는 정수, 실수와 관계없이 0이면 거짓, 0이 아니면 참임)
if 0:
    print('참')
else:
    print('거짓')

if 1:
    print('참')
else:
    print('거짓')

if 0x1F:
    print('참')
else:
    print('거짓')

if 0b1000:
    print('참')
else:
    print('거짓')

if 13.5:
    print('참')
else:
    print('거짓')
```

```python
#  조건문에 문자열 지정하기 (문자열은 내용이 있으면 참, 없으면 거짓)
if 'hello':
    print('참')
else:
    print('거짓')

if ' ':              # 공백이라는 내용이 있는 것
    print('참')
else:
    print('거짓')

if '':
    print('참')
else:
    print('거짓')
```

```python
if not 0:     # 0은 Falss 의 not --> True
    print('참')

if not None:    # None은 False의 not --> True
    print('참')

if not "":      # 빈 문자열은 False의 not --> True
    print('참')
```

```python
# 조건식을 여러 개 지정하기
x = 10
y = 20

if x == 10 and y == 20:
    print('참')
else:
    print('거짓')
```

```python
x = 15

if x > 0 and x < 20:
    print('입장 가능합니다')
if 0 < x < 20:
    print('입장 가능합니다')
if x > 0:
    if x < 20:
        print('입장 가능합니다.')   # 위의 세 가지가 모두 같은 것임


y = 70
if y < 8 or y > 65:
    print('할인 대상입니다.')
```

```python
# 문자열 조작하기
## 문자열 바꾸기

print("Hello, World".replace("World", "Python"))

table = str.maketrans("aeiou", "12345")   # aeiou 를 12345로 변환하여 table에 저장하고,
print("apple".translate(table))           # apple 문자를 table에 저장한 값(aeiou --> 12345)로 바꿈
```

```python
## 구분자 문자열과 문자열 리스트 연결하기
lis = ['apple', 'pear', 'grape', 'pineapple', 'arange']

" ".join(lis)
"".join(lis)
"-".join(lis)
```

```python
## 소문자를 대문자로, 대문자를 소문자로 바꾸기
print("Python".upper())
print("Python".lower())
```

```python
## 공백 삭제하기
print('   Python   '.strip())
print('   Python   '.lstrip())
print('   Python   '.rstrip())
```

```python
## 특정 문자 삭제하기
print(',  python.'.strip(',.'))
print(',  python.'.lstrip(',.'))
print(',  python.'.rstrip(',.'))
print(', !python.!'.strip(',.!'))   # 양 끝에 있는 , 와 . 과 ! 만 지울 수 있음
print(', !python.!'.strip(',.! '))   # 양 끝에 있는 , 와 . 과 !와 공백 지울 수 있음
print('*, !python.!'.strip(',.! '))   # 끝에 있는 게 별이기 때문에 안지워짐
```

```python
## 문자열을 렬하기
print("python".ljust(10))  # 문자 길이를 10으로 지정하고 나머지는 공백, 즉, 문자열 길이 6 + 공백 4
print("python".rjust(10))  # 문자 길이를 10으로 지정하고 나머지는 공백, 즉, 공백 4 + 문자열  길이 6
print("python".center(10))  # 공백 2 + 문자열 길이 6 + 공백 2
```

```python
## 메서드 체이닝  ( 메서드를 줄줄이 연결한다는 뜻)
print("python".rjust(10).upper())
print("python".rjust(10).upper().strip())
```

```python
## 문자열 왼쪽에 0 채우기
print("35".zfill(4)) # 4칸을 확보한 상태에서 앞에 2 칸이 남아서 0으로 채움
print("3.5".zfill(6)) # 소수점도 한 칸임
```

```python
## 문자열 위치 찾기
print("apple pineapple".find("pl"))
print("apple pineapple".find("xy"))   # 없는 문자열을 찾으려고 하면 -1이 나옴
```

```python
"apple pineapple".rfind("pl")  # 오른쪽에서 찾기
```

```python
# 문자열 서식 지정자와 포매팅 사용하기
## 서식 지정자로 문자열 넣기
# "%s" % "문자열"
print("I am %s" % "James")

name = "Maria"
print("I am %s" % name)
```

```python
# 서식 지정자로 숫자 넣기
# "%d" % 숫자
print("I am %d years old" % 28)  # %d는 int
```

```python
# 서식 지정자로 소수점 표현하기
# "%f" % 숫자
print("%f" % 2.3)    # %f는 float

# "%.자리수f" % 숫자
print("%.2f" % 2.3)   # 소수점 뒤 두자리
print("%.3f" % 2.3)   # 소수점 뒤 세자리
print("%10.3f" % 2.3)  # 10칸을 확보하고 소수점 뒤 세 자리
```

```python
## 서식지정자로 문자열 정렬하기
print("%10s" % "python")  # 문자열 길이를 10으로 만들고, 문자를 오른쪽 정렬함 --> 공백 4칸 + 문자 6칸
print("%-10s" % "python")  # 문자열 길이를 10으로 만들고, 문자를 왼쪽 정렬함 --> 문자 6칸 + 공백 4칸
```

```python
## 서식지정자로 문자열 안에 값 여러 개 넣기
print("Today is %d %s" % (3, "April"))
```

```python
## format 메서드 사용하기
"hello {0}".format("world")
```

```python
# 채우기와 정렬을 조합해서 사용하기
# '{인덱스:[[채우기]정렬][길이][.자릿수][자료형]}'
print('{0:<10}'.format('Python'))  # 길이10, 왼쪽으로 정렬
print('{:>10}'.format('Python'))  # 길이10, 오른쪽으로 정렬
```

```python
# 숫자 개수 맞추기
print('%03d' % 1)  # 자리 수에 맞춰 0 을 널음, 3자리수에 남는 자리는 0으로 채움
print('{0:03d}'.format(35))
print('%008.2f' % 3.6)    # 8자리를 확보하고 소수점 이하 2자리, 나머지는 0으로 채움
print('{0:08.2f}'.format(3.6))   # 8칸을 확보하고 비어있는 공간에 0을 주고 소수점은 2자리
```

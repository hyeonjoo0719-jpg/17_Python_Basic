```python
# 문자열 조작하기
## 문자열 바꾸기

"Hello, World".replace("World", "Python")
```

```python
## 문자바꾸기

table = str.maketrans("aeiou", "12345")
"apple".translate(table)
```

```python
## 문자열 분리하기
text = "apple pear grape pineapple arange"
text.split()
```

```python
text = "apple, pear, grape, pineapple, arange"
text.split(", ")
```

```python
## 구분자 문자열과 문자열 리스트 연결하기
lis = ['apple', 'pear', 'grape', 'pineapple', 'arange']

" ".join(lis)
"".join(lis)
"-".join(lis)
```

```python
## 소문자를 대문자로, 대문자를 소문자로 바꾸기
print("Python".upper())
print("Python".lower())
```

```python
## 공백 삭제하기
print('   Python   '.strip())
print('   Python   '.lstrip())
print('   Python   '.rstrip())
```

```python
## 특정 문자 삭제하기
print(',  python.'.strip(',.'))
print(',  python.'.lstrip(',.'))
print(',  python.'.rstrip(',.'))
print(', !python.!'.strip(',.!'))   # 양 끝에 있는 , 와 . 과 ! 만 지울 수 있음
print(', !python.!'.strip(',.! '))   # 양 끝에 있는 , 와 . 과 !와 공백 지울 수 있음
print('*, !python.!'.strip(',.! '))   # 끝에 있는 게 별이기 때문에 안지워짐
```

```python
## 문자열을정렬하기
print("python".ljust(10))
print("python".rjust(10))
print("python".center(10))  # 양쪽에 공백 10칸을 확보
```

```python
## 메서드 체이닝
print("python".rjust(10).upper())
print("python".rjust(10).upper().strip())
```

```python
## 문자열 왼쪽에 0 채우기
print("35".zfill(4)) # 4칸을 확보한 상태에서 앞에 2 칸이 남아서 0으로 채움
print("3.5".zfill(6))
```

```python
## 문자열 위치 찾기
print("apple pineapple".find("pl"))
print("apple pineapple".find("xy"))   # 없는 문자열을 찾으려고 하면 -1이 나옴
```

```python
"apple pineapple".rfind("pl")  # 오른쪽에서 찾기
```

```python
# 문자열 서식 지정자와 포매팅 사용하기
## 서식 지정자로 문자열 넣기
# "%s" % "문자열"
print("I am %s" % "James")

name = "Maria"
print("I am %s" % name)
```

```python
# 서식 지정자로 숫자 넣기
# "%d" % 숫자
print("I am %d years old" % 28)
```

```python
# 서식 지정자로 소수점 표현하기
# "%f" % 숫자
print("%f" % 2.3)

# "%.자리수f" % 숫자
print("%.2f" % 2.3)
print("%.3f" % 2.3)
print("%10.3f" % 2.3)
```

```python
## 서식지정자로 문자열 정렬하기
print("%10s" % "python")
print("%-10s" % "python")
```

```python
## 서식지정자로 문자열 안에 값 여러 개 넣기
print("Today is %d %s" % (3, "April"))
```

```python
## format 메서드 사용하기
"hello {0}".format("world")
```

```python
"Hello {0} {1} {2}".format("Python", "Script", 3.6)
```

```python
"{0} {0} {1} {1}".format("python", "script")
```

```python
"Hello {language} {version}".format(language="python", version=3.6)
```

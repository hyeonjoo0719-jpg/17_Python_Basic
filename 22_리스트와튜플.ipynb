```python
# 리스트의 할당(참조)
a = [0, 0, 0, 0, 0]
b = a

print(a == b)
print(a is b)
print(id(a))
print(id(b))

b[2] = 99   # 같은 주소를 참조하고 있기 때문에 똑같이 변함
print(a)
print(b)
```

```python
# 리스트의 복사
a = [0, 0, 0, 0, 0]
b = a.copy()

print(a)
print(b)

print(a == b)   # 값은 같지만 
print(a is b)   # 주소는 다름

print(id(a))
print(id(b))

b[3] = 99
print(a)    # 주소는 다르기 때문에 b를 바꿔도 a는 바뀌지 않음
print(b)
```

```python
# 인덱스와 요소 함께 출력하기 - enumerate()
a = [38, 21, 53, 62, 19]
for value in a:
    print(value)

a = [38, 21, 53, 62, 19]
for value in enumerate(a):   # 인덱스 값과 함께 튜플로 출력해줌
    print(value)

for index, value in enumerate(a):   
    print(f'{index+1}번째 값은 {value}입니다.')
```

```python
# while문으로 요소 출력하기
a = [38, 21, 53, 62, 19]

i = 0
while i < len(a):
    print(a[i])
    i += 1
```

```python
# 리스트의 가장 작은 수와 가장 큰 수와 합께 구하기
# min(), max(), sum()
a = [38, 21, 53, 62, 19]
smallest = a[0]

for i in a:
    if i < smallest:          
        smallest = i          #가장 작은 수를 i로 재할당 하기
print(smallest)
```

```python
a = [38, 21, 53, 62, 19]
largest = a[0]

for i in a:
    if i > largest:
        largest = i          #가장 큰 수를 i로 재할당 하기
print(largest)
```

```python
a = [38, 21, 53, 62, 19]
x = 0

for i in a:
    x = x + i     # 총합 구하기(=sum)

print(x)
```

```python
# 리스트 표현식 사용하기
# [식 for 변수 in 리스트]
# list(식 for 변수 in 리스트)

a = [i for i in range(10)]
print(a)

b = list(i for i in range(10))
print(b)
```

```python
# 리스트 표현식에서 if 조건문 사용하기

# [식(변수) for 변수 in 리스트 if 조건식]
# list(식(변수) for 변수 in 리스트 if 조건식)

a = [i for i in range(10) if i % 2 == 0]
print(a)

print()         # 위와 아래는 같은 코드

a = []
for i in range(10):
    if i % 2 == 0:
        a.append(i)
print(a)
```

```python
b = [ i + 5 for i in range(10) if i % 2 == 1]
print(b)
```

```python
# for 반복문과 if 조건문을 여러 번 사용하기
a = [i * j for i in range(2, 10) for j in range(1, 10)]
print(a)

print()    # 위와 아래는 같음

b = []
for i in range(2, 10):
    for j in range(1, 10):
        b.append(i * j)
print(b)
```

```python
# 리스트에 map 사용하기
# list(map(함수, 리스트))
# tuple(map(함수, 튜플))

a = [1.2, 2.5, 3.7, 4.6]

for i in range(len(a)):
    a[i] = int(a[i])
print(a)
```

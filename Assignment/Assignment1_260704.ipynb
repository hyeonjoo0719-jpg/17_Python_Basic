```python
# 문제 1
students = {
    '김철수': [90, 80, 70],
    '이영희': [100, 95, 90],
    '박민수': [60, 70, 65]
}

total_average = 0

print("===== 학생 성적 결과 =====\n")
for name, score in students.items():
    average = sum(score) / len(score)
    total_average += average
    print(f"이름: {name}\n")
    print(f"점수: {score}\n")
    print(f"평균: {average}\n")
    print()

print("===== 전체 결과 =====\n")
print(f"전체 학생 수: {len(students)}명\n")
print(f"전체 평균: {total_average / len(students)}")
```

```python
#문제 7
height = int(input())

for i in range(height):
    for j in range(i):       # 공백 출력
        print(" ", end="")
    for k in range(2 * (height - i) -1):   # 별 출력
        print('*', end='')
    print()
```

```python
#문제 9
## 일반표현식
num1, num2 = map(int, input().split())
num_list = []

for i in range(num1, num2 +1):
    num_list.append(round(i ** 0.5, 1))

print(num_list)
```

```python
## 컴프리헨션
num_list = [round( i ** 0.5, 1) for i in range(num1, num2 +1)]
print(num_list)
```

```python
# 문제 10
matrix = []
row, col = map(int, input().split())

for i in range(row):
    matrix.append(list(input()))

for r in range(row):
    for c in range(col):
        if matrix[r][c] == '*':
            continue

        count = 0
        for nr in range(r-1, r+2):
            for cr in range(c-1, c+2):

                if 0 <= nr < row and 0 <= cr < col:
                    if matrix[nr][cr] == '*':
                        count += 1
        matrix[r][c] = str(count)

for r in range(row):
    print(''.join(matrix[r]))
```

```python
result = 0  # 전역변수

def add(num):
    global result    # 전역변수
    result = result + num
    return result

print(add(3))
print(add(4))
print(add(5))
```

```python
result1 = 0  # 전역변수

def add(num):
    global result1    # 전역변수
    result1 += num
    return result1

print(add(1))
print(add(3))
print(add(4))
```

```python
result1 = 0
result2 = 0

def add1(num):
    global result1    # 전역변수
    result1 += num
    return result1

def add2(num):
    global result2    # 전역변수
    result2 += num
    return result2

print(add1(1))
print(add1(3))

print(add2(5))
print(add2(6))
```

```python
class Calculator:
    def __init__(self):        # 초기값 세팅
        self.result = 0
    def add(self, num):
        self.result += num
        return self.result

cal1 = Calculator()
print(cal1.add(3))
print(cal1.add(4))

cal2 = Calculator()
print(cal2.add(5))
print(cal2.add(6))
```

```python
# 속성 사용하기
# class 클래스이름:
#     def __init__(self):
#         self 속성 = 값

class Person:
    def __init__(self):
        self.hello = '안녕하세요'

    def greeting(self):
        print(self.hello)

james = Person()
james.greeting()
```

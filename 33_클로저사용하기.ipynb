```python
# 함수 안에서 전역변수 변경하기

b = 10
def foo():
    b = 20
    print(b)
foo()
print(b)
```

```python
c = 10
def foo():
    global c          # c를 전역변수로 쓰겠다고 선언, 해당 로직을 나가도 변수로 기능함
    c = 20
    print(c)
foo()
print(c)
```

```python
def foo():
    global d           
    d = 20
    print(d)
foo()
print(d)          # c를 전역변수로 쓰겠다고 선언했기 때문에 , 해당 로직을 나가도 변수로 기능함, (원래 d 라는 인수가 정의되지 않았지만 foo 로직 안에서 정의되었음)
```

```python
# 함수 안에서 함수 만들기

# def 함수이름():
# def 함수이름():
#     코드
#     def 함수이름():
#         코드

def print_hello():
    hello = 'Hello, world!'
    def print_massage():
        print(hello)
    print_massage()
print_hello()
```

```python
# 지역 변수 변경하기
def A():
    x = 10
    def B():
        x = 20
    B()
    print(x)
A()
```

```python
def A():
    x = 10
    def B():
        nonlocal x       # 전역변수도, 지역변수도 아님,, 제일 가까운 변수를 활용
        x = 20
    B()
    print(x)

A()
```

```python
# nonlocal이 변수를 찾는 순서

def A():
    x = 10
    y = 100

    def B():
        x = 20
        def C():
            nonlocal x
            nonlocal y
            x = x + 30
            y = y + 300
            print(x)
            print(y)
        C()
    B()
A()
```

```python
# global로 전역 변수 사용하기

x = 1
def A():
    x = 10
    def B():
        x = 20
        def C():
            global x       # 글로벌이 있기에 가장 바깥에 있는 전역변수 x = 1이 됨
            x = x + 30
            print(x)
        C()
    B()
A()
```

```python
# 클로저 사용하기
def calc():
    a = 3
    b = 5
    def mul_add(x):
        return a * x + b
    return mul_add     # 괄호가 없고 함수명만 있음(바로 실행 하지 않고 함수명만 가지고 옴)

c = calc()  # 3x + 5
print(c(1), c(2), c(3))
```

```python
# lambda로 클로저 만들기
def calc():
    a = 3
    b = 5
    return lambda x : a * x + b

c = calc()
print(c(4))
```

```python
#33.5
def counter():
    i = 0
    def count():
        nonlocal i
        i += 1
        return i
    return count
c = counter()
for i in range(10):
    print(c(), end=' ')
```

```python
# for 변수 in range(횟수):
#     반복할 코드

for i in range(10):
    print(f'{i} Hello World')
```

```python
for _ in range(10):      # 10번만 반복하고 변수를 안쓰려면 _ 사용
    print(f'Hello World')
```

```python
# for와 range 응용하기
for i in range(5, 12):
    print('Hello world', i)
```

```python
for a in range(0, 10, 2):
    print('hello world', a)
```

```python
# 숫자를 감소시키기
for j in range(10, 0, -1):
    print('Hello world!', j)
print()
for a in reversed(range(10)):
    print('Hello world!', a)
```

```python
# 입력한 횟수대로 반복하기
count = int(input('반복할 횟수를 입력하세요: '))

for i in range(count):
    print('hello world', i)
```

```python
# 시퀀스 객체로 반복하기
a = [10, 20, 30, 40, 50]
for i in a:
    print(i)
```

```python
## 튜플
fruits = (
    'apple', 'orange', 'grape'
)
for fruit in fruits:
    print(fruit)
```

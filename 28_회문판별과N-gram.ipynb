```python
# 회문판별하기
## 반복문으로 문자 검사하기

word = input("단어를 입력하세요: ")

is_palindrome = True

for i in range(len(word) //2 ):
    if word[i] != word[-1 -i]: 
        is_palindrome = False      # 같지 않으면 False로 바꿔서 반복문을 끝냄
print(is_palindrome)
```

```python
## 시퀀스 뒤집기로 문자 검사하기
word = input('단어를 입력하세요: ')

print(word == word[::-1])
```

```python
#3 리스트와 reversed 사용하기
word = 'level'
list(word) == list(reversed(word))
```

```python
## 문자열의 join 메서드와 reversed 사용하기
word = 'level'
''.join(reversed(word))
```

```python
# N-gram 만들기(N = 출력할 문자 수)
## 반복문으로 N-gram 출력하기

text = "Hello"

for i in range(len(text)-1):
    print(text[i], text[i+1], sep='')
```

```python
text = 'this is python script'
words = text.split()

for i in range(len(words) -1):
    print(words[i], words[i +1])
```

```python
## zip으로 2-gram 만들기
text = "hello"
two_gram = zip(text, text[1:])
# list(two_gram)
# [('h', 'e'), ('e', 'l'), ('l', 'l'), ('l', 'o')]

for i in two_gram:
    print(i[0], i[1], sep='')
```

```python
## zip과 리스트 표현식으로 N-gram 만들기
text = 'hello'
[ text[i:] for i in range(3)]
```

```python
list(zip(*['hello', 'ello', 'llo']))
```

```python
#28.3
n = int(input())
text = input()
word = text.split()

if len(word) < n:
    print('wrong')
else:
    n_gram = zip(*[text[i:] for i in range(n)])
    for i in n_gram:
        print(i)
```

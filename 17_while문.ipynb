```python
# while 반복문 사용하기
# 초기식
# while 조건식:
#     반복할 코드
#     변화식
# i = 0
# while i < 10:
#     print('hello world')
#     i += 1

for j in range(10):
    print('hello world')
```

```python
# 초기값을 1부터 시작하기
i = 1
while i <= 10:
    print('Hello world', i)
    i += 1
```

```python
# 입력한 횟수대로 반복하기
count = int(input('반복할 횟수를 입력하세요'))

i = 0
while i < count:
    print('Hello World', i)
    i += 1
print()

# 초깃값을 받은만큼 출력
count2 = int(input('반복할 횟수를 입력하세요: '))

while count > 0:
    print('Hello world', count)
    count -= 1
```

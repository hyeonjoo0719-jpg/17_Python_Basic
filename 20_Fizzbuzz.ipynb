```python
# 3의 배수일 때와 5의 배수일 때 처리하기
for i in range(1, 21):
    if i% 3 == 0:
        print('fizz')
    elif i % 5 == 0:
        print('Buzz')
    else:
        print(i)
```

```python
# 3과 5의 공배수 처리하기
for i in range(1, 21):
    if i % 3 == 0 and i % 5 == 0:
        print('Fizzbuzz')
    elif i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    else:
        print(i)

print()   # 위 아래 같은 코드임

for i in range(1, 21):
    if i % 15 == 0:
        print('Fizzbuzz')
    elif i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    else:
        print(i)
```

```python
# 코드 단축하기
for i in range(1, 21):
    print('fizz' * (i % 3 == 0) + 'Buzz' * ( i % 5 == 0) or i)  # False는 0음 --> 1이면 fizz * 0 + buzz * 0 + 1이 돼서 1이 출력됨
                                                                # 3은 3 % 3 == 0 이라서 True가 돼서 fizz * 1 이라서 fizz가 출력됨
```

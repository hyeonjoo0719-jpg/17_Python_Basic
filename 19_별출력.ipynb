```python
for i in range(5):           # 바깥쪽 루프는  세로 방향
    for j in range(5):       # 안쪽 루프는 가로 방향
        print("j:", j, sep='', end=' ')
    print('i: ', i, '\\n', sep='')
```

```python
# 사각형으로 별 출력하기
for i in range(5):
    for j in range(5):
        print('*', end='')
    print()
```

```python
for i in range(3):
    for j in range(7):
        print('*', end='')
    print()
```

```python
# 계단식으로 별 출력하기
for i in range(5):
    for j in range(5):
        if j <= i:
            print('*', end='')
    print()
```

```python
# 대각선으로 별 출력하기
for i in range(5):
    for j in range(5):
        if j == i:
            print('*', end='')
        else:
            print(' ', end='')
    print()
```

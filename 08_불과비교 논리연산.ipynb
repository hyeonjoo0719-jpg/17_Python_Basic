```python
# 비교연산자
3 > 1

# 숫자가 같은지 다른지
print(10 == 10)
print(10 != 10)
print()

# 문자열이 같은지
print('Python' == 'Python')
print('Python' == 'python')
print('Python' != 'python')
print()

# 부등호 사용하기
print(10 > 20)
print(10 < 20)
print(10 >= 20)
print(10 <= 20)
print()

# 객체가 같은지 다른지
print(1 == 1.0)
# print(1 is 1.0)
print()

# 논리 연산자 사용하기
print(True and True)
print(True and False)
print(False and True)
print(False and False)
print()
print(True or True)
print(True or False)
print(False or True)
print(False or False)
print()

print(not True)
print(not False)
```

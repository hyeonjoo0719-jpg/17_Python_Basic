```python
# 딕셔너리 조작하기
## setdefault, update (매서드를 이용해서 추가하기)

a = {
    'a': 10,
    'b': 20,
    'c': 30,
    'd': 40
}

a.setdefault('e')
print(a)

a.setdefault('f', 100)
print(a)
```

```python
## 딕셔너리의 키 값 수정하기
a = {
    'a': 10,
    'b': 20,
    'c': 30,
    'd': 40
}

a.update(a=90)
print(a)

a.update(e=90)
print(a)

a.update(a=900, f=60)  # 있는 a는 900으로  수정, 없는 f는 추가
print(a)
```

```python
y = {
    1: 'one',
    2: 'two'
}

y.update({1:'ONE',2: 'TWO'})
print(y)
```

```python
y.update(
    [
        [2, 'TWO'],        # 2차원 배열
        [4, "FOUR"]
    ]
)
print(y)
```

```python
y.update(
    zip([1, 2], ['one', 'two'])
)
print(y)
```

```python
## 딕셔너리에서 키-값 쌍 삭제하기
x = {
    'a': 10,
    'b': 20,
    'c': 30,
    'd': 40
}

result = x.pop('a')    # value 값만 뽑아냄

print(result)
print(x)

x.pop('z', 0)   # 0을 안적으면 에러가 나고, 0을 적으면 없으면 0을 반환함

del x['b']
print(x)
```

```python
## 딕셔너리에서 임의의 키-값 쌍 삭제하기
x = {
    'a': 10,
    'b': 20,
    'c': 30,
    'd': 40
} 
result = x.popitem()    # 기본으로 -1이 입력되어 있어 가장 마지막의 키-값 쌍을 튜플 형태로 뽑아냄
print(result)
print(x)

x.clear()
print(x)
```

```python
## 딕셔너리에서 키-값 쌍을 모두 가져오기
x = {
    'a': 10,
    'b': 20,
    'c': 30,
    'd': 40
} 
print(x.items()) 
print(x.keys())
print(x.values())
```

```python
## 리스트와 튜플로 딕셔너리 만들기
keys = ['a', 'b', 'c', 'd']
x = dict.fromkeys(keys)  # {}.fromkeys와 같음,  키 리스트로 딕셔너리를 생성하며 값은 모두 None으로 저장
print(x)

y = {}.fromkeys(keys, 100)   # 키 리스트로 딕셔너리를 생성하며 값은 모두 100으로 저장
print(y)
```

```python
## 반복문으로 딕셔너리의 키-값 쌍을 모두 출력하기
x = {
    'a': 10,
    'b': 20,
    'c': 30,
    'd': 40
} 
for i in x:
    print(i)   # 키 값만 가지고옴

for i in x.items():
    print(i)   # 키-값 쌍 모두 가지고 옴

for key, value in x.items():
    print(key, value)
```

```python
# 키만 출력하기
for key in x.keys():
    print(key, end=' ')

print()

# value만 출력하기
for value in x.values():
    print(value, end=' ')
```

```python
## 딕셔너리 컴프리헨션

# {키:값 for 키, 값 in 딕셔너리}
# dict({키:값 for 키, 값 in 딕셔너리})

keys = ['a', 'b', 'c', 'd']

x = {key: value for key, value in dict.fromkeys(keys).items()}
print(x)
print()

imsi = {
    'a': None,
    'b': None,
    'c': None,
    'd': None
}
y = {key: value for key, value in imsi.items()}
print(y)
```

```python
print({key : 0 for key in dict.fromkeys(['a', 'b', 'c', 'd']).keys()})
print({value : 0 for value in {'a':10, 'b':20, 'c':30, 'd':40}.values()})
```

```python
## 딕셔너리 표현식에서 if 조건문 사용하기
x = {
    'a': 10,
    'b': 20,
    'c': 30,
    'd': 40
} 
for key, value in x.items():
    if value == 20:
        del x[key]            # 키와 값의 쌍을 삭제는 안됨

print(x)
```

```python
x = {
    'a': 10,
    'b': 20,
    'c': 30,
    'd': 40
} 
x = {key:value for key, value in x.items() if value != 20}  # value가 20이 아닐 때만 쌍을 리스트에 추가하는 형식
print(x)
```
